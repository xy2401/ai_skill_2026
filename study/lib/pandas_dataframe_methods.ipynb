### Help for `apply`

```python
help(pandas.DataFrame.apply)
```

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})
print(df['A'].apply(lambda x: x**2))
```

### Help for `describe`

```python
help(pandas.DataFrame.describe)
```

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4, 5]})
print(df.describe())
```

### Help for `dropna`

```python
help(pandas.DataFrame.dropna)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A': [1, np.nan, 3]})
print(df.dropna())
```

### Help for `fillna`

```python
help(pandas.DataFrame.fillna)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A': [1, np.nan, 3]})
print(df.fillna(0))
```

### Help for `groupby`

```python
help(pandas.DataFrame.groupby)
```

```python
import pandas as pd
df = pd.DataFrame({'Key': ['A', 'B', 'A', 'B'], 'Val': [1, 2, 3, 4]})
print(df.groupby('Key').sum())
```
