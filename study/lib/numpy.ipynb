### Help for `argmax`

```python
help(numpy.argmax)
```

```python
import numpy as np
arr = np.array([1, 5, 2])
print(f'Index of max element: {np.argmax(arr)}')
```

### Help for `concatenate`

```python
help(numpy.concatenate)
```

```python
import numpy as np
a = np.array([1, 2])
b = np.array([3, 4])
print(f'Concatenated: {np.concatenate((a, b))}')
```

### Help for `expand_dims`

```python
help(numpy.expand_dims)
```

```python
import numpy as np
a = np.array([1, 2])
expanded = np.expand_dims(a, axis=0)
print(f'Original shape: {a.shape}, Expanded shape: {expanded.shape}')
```
