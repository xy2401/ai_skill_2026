### Help for `LabelEncoder`

```python
help(sklearn.preprocessing.LabelEncoder)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
print(le.fit_transform(['tokyo', 'paris', 'tokyo']))
```

### Help for `MinMaxScaler`

```python
help(sklearn.preprocessing.MinMaxScaler)
```

```python
from sklearn.preprocessing import MinMaxScaler
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = MinMaxScaler()
print(scaler.fit_transform(data))
```

### Help for `StandardScaler`

```python
help(sklearn.preprocessing.StandardScaler)
```

```python
from sklearn.preprocessing import StandardScaler
data = [[0, 0], [0, 0], [1, 1], [1, 1]]
scaler = StandardScaler()
print(scaler.fit_transform(data))
```
