```python
import os
try:
    # We create a dummy image first
    img = PIL.Image.new('RGB', (100, 100), color = 'red')
    img.save('test.png')
    img2 = PIL.Image.open('test.png')
    print(f'Opened image size: {img2.size}')
    os.remove('test.png')
except Exception as e:
    print(e)
```
