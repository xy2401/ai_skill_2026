```python
import numpy as np
import cv2

# 1. 创建画布：高度(H)=100, 宽度(W)=150, 通道(C)=3
# 【核心预警】：NumPy 的顺序是 (行, 列) -> 对应 (高度, 宽度)
# 这里我们故意让 宽 > 高，这样 shape 就会显示为 (100, 150, 3)
img = np.zeros((100, 150, 3), dtype=np.uint8)

print(f'原始彩色图 shape: {img.shape}') 
# 输出: (100, 150, 3) -> [H, W, C]

# 2. 转换为灰度图
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. 观察转换后的维度
# 【核心发现】：第 3 个维度（通道轴）被“拍扁”了
print(f'转换后灰度图 shape: {gray.shape}') 
# 输出: (100, 150) -> 只有 [H, W]
```

```python
import os
import cv2
import numpy as np

# 1. 在内存中创建一个“假”图像（Dummy Image）
# 创建一个 10x10 像素、3 通道（BGR）的全黑矩阵
# dtype=np.uint8 必须明确指定，因为 OpenCV 要求图像像素为 8 位无符号整数（0-255）
dummy_data = np.zeros((10, 10, 3), dtype=np.uint8)

# 2. 持久化：将内存中的数组保存为本地文件
# cv2.imwrite 会根据文件后缀名（.png）自动调用相应的压缩编码器
cv2.imwrite('test_cv.png', dummy_data)

# 3. 读取：从磁盘加载刚才保存的图片
# 此时 img 又变回了一个 NumPy 数组
img = cv2.imread('test_cv.png')

# 4. 验证：检查读取后的维度信息
# 如果文件读取失败（比如路径错或文件损坏），img 会是 None，访问 .shape 会报错
if img is not None:
    print(f'Read image shape: {img.shape}') # 输出: (10, 10, 3)
else:
    print("读取失败，请检查文件是否存在！")

# 5. 清理：删除临时产生的物理文件
# 在自动化测试或批量处理中，及时删除临时文件可以防止磁盘空间被占满
if os.path.exists('test_cv.png'):
    os.remove('test_cv.png')
    print("临时文件已成功删除。")
```

```python
import cv2
import numpy as np
import os

# 定义含中文的路径
path_cn = "图像_测试_2026.jpg"

# --- 【读取】中文路径图片 ---
def imread_cn(file_path):
    # np.fromfile 从磁盘读取原始二进制数据，dtype=np.uint8 是图像的标准格式
    # 这步不经过 OpenCV，所以不受中文路径限制
    raw_data = np.fromfile(file_path, dtype=np.uint8)
    
    # cv2.imdecode 将二进制流解码为图像矩阵
    # cv2.IMREAD_COLOR 确保读入的是彩色 BGR 格式
    img = cv2.imdecode(raw_data, cv2.IMREAD_COLOR)
    return img

# --- 【保存】中文路径图片 ---
def imwrite_cn(file_path, img):
    # .jpg 是你想要保存的格式，img 是你的 NumPy 数组图片
    # cv2.imencode 返回一个元组 (成功标志, 编码后的二进制流)
    ext = os.path.splitext(file_path)[-1] # 自动获取后缀，如 .png 或 .jpg
    success, encoded_img = cv2.imencode(ext, img)
    
    if success:
        # 将二进制流写入文件
        encoded_img.tofile(file_path)
        return True
    return False

# --- 测试运行 ---
# 1. 创建一个蓝色方块
test_img = np.zeros((200, 200, 3), dtype=np.uint8)
test_img[:] = [255, 0, 0]  # BGR 蓝色

# 2. 保存到中文路径
if imwrite_cn(path_cn, test_img):
    print(f"成功保存到: {path_cn}")

# 3. 从中文路径读取
read_img = imread_cn(path_cn)
if read_img is not None:
    print(f"成功读取，尺寸为: {read_img.shape}")

# 清理测试文件
# os.remove(path_cn)
```
