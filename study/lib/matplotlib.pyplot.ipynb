matplotlib.pyplot 是一组函数的集合，使得 Matplotlib 的工作方式类似于 MATLAB。

```python
import matplotlib.pyplot
print(f'matplotlib.pyplot version: {matplotlib.pyplot.__version__ if hasattr(matplotlib.pyplot, "__version__") else "N/A"}')
```

```python
help(matplotlib.pyplot)
```

```python
dir(matplotlib.pyplot)
```

### Help for `cm`

```python
help(matplotlib.pyplot.cm)
```

```python
import matplotlib.pyplot as plt
print(f'Color from viridis: {plt.cm.viridis(0.5)}')
```

### Help for `figure`

```python
help(matplotlib.pyplot.figure)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(4, 2))
print(f'Created figure: {fig}')
```

### Help for `legend`

```python
help(matplotlib.pyplot.legend)
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2], [1, 2], label='Test')
plt.legend()
print('Legend added')
```

### Help for `scatter`

```python
help(matplotlib.pyplot.scatter)
```

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3], [4, 5, 6])
print('Scatter plot created')
```

### Help for `show`

```python
help(matplotlib.pyplot.show)
```
