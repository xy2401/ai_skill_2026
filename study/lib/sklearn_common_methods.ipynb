### Help for `fit`

```python
# Help for common method 'fit' is usually found in specific classes.
# Here we show a generic help or example.
print('Scikit-learn fit method')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.dot(X, np.array([1, 2])) + 3
reg = LinearRegression().fit(X, y)
print(f'Model fitted. Coef: {reg.coef_}')
```

### Help for `fit_transform`

```python
# Help for common method 'fit_transform' is usually found in specific classes.
# Here we show a generic help or example.
print('Scikit-learn fit_transform method')
```

```python
from sklearn.preprocessing import StandardScaler
data = [[0, 0], [0, 0], [1, 1], [1, 1]]
scaler = StandardScaler()
print(f'Transformed data:\n{scaler.fit_transform(data)}')
```

### Help for `predict`

```python
# Help for common method 'predict' is usually found in specific classes.
# Here we show a generic help or example.
print('Scikit-learn predict method')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.array([[1, 1], [1, 2]])
y = [3, 4]
reg = LinearRegression().fit(X, y)
print(f'Prediction: {reg.predict([[2, 2]])}')
```

### Help for `score`

```python
# Help for common method 'score' is usually found in specific classes.
# Here we show a generic help or example.
print('Scikit-learn score method')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.array([[1, 1], [1, 2]])
y = [3, 4]
reg = LinearRegression().fit(X, y)
print(f'Score: {reg.score(X, y)}')
```
